## 检查naive队列

```python
info=pd.read_excel('/work/longyh/BY/processed/merged_WES_meta_cohort_naive.xlsx')
```

```python
info.head()
```

```python
info['Finish'].fillna(0,inplace=True)
```

```python
#获取Finish为0的Run，存储为list
run_list=info[info['Finish']==0]['Run'].tolist()
```

```python
#查看target_dir是否存在run_list中对应的fastq文件
target_dir='/work/longyh/BY/fastq/WES/'
existing_files=os.listdir(target_dir)
```

```python

# 筛出 run_list 中未在目标目录出现的条目（文件名包含 run ID 即视为存在）
missing_runs = [
    run for run in run_list
    if not any(run in fname for fname in existing_files)
]

print(f"目标目录共 {len(existing_files)} 个文件/文件夹")
print(f"待检查 run 数：{len(run_list)}")
print(f"未找到的 run 数：{len(missing_runs)}")

if missing_runs:
    print("缺失的 run：")
    for r in missing_runs:
        print(r)
else:
    print("所有 run 对应的 fastq 均已存在。")
```

```python
len(missing_runs)
```

```python
#保存missing_runs
with open('/work/longyh/BY/processed/missing_WES_run_naive.txt', 'w') as f:
    for run in missing_runs:
        f.write(f"{run}\n")
```

## 查看有RNA-seq的naive患者有多少

```python
naive_path='/work/longyh/BY/processed/merged_WES_meta_cohort_naive.xlsx'
naive=pd.read_excel(naive_path)
```

```python
naive.head()
```

```python
seq_table=pd.read_excel('/work/longyh/BY/raw/1-s2.0-S0092867417311224-mmc4.xlsx',skiprows=2)
```

```python
seq_table.head()
```

```python
#naive队列取Patient列并去重
naive_patients = naive["Patient"].drop_duplicates()
```

```python
merged_naive = pd.merge(naive_patients.to_frame(), seq_table, on="Patient", how="left")
```

```python
merged_naive.head()
```

```python
merged_naive.shape
```

```python
#Pre-treatment Exome和Pre-treatment RNA-Seq都为1的患者数量
pre_treatment_both_1 = merged_naive[(merged_naive["Pre-treatment Exome"] == 1) & (merged_naive["Pre-treatment RNA-Seq"] == 1)]
pre_treatment_both_1.shape[0]
```

```python
#Pre-treatment Exome和On-treatment RNA-Seq都为1的患者数量
pre_treatment_exome_on_treatment_rna_1 = merged_naive[(merged_naive["Pre-treatment Exome"] == 1) & (merged_naive["On-treatment RNA-Seq"] == 1)]
pre_treatment_exome_on_treatment_rna_1.shape[0]
```

```python
#Pre-treatment Exome为1且Pre-treatment RNA-Seq或On-treatment RNA-Seq为1的患者数量
pre_treatment_exome_post_on_treatment_rna_1 = merged_naive[(merged_naive["Pre-treatment Exome"] == 1) & ((merged_naive["Pre-treatment RNA-Seq"] == 1) | (merged_naive["On-treatment RNA-Seq"] == 1))]
pre_treatment_exome_post_on_treatment_rna_1.shape[0]
```
