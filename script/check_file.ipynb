WES的文件下载信息

```python
WES_meta=pd.read_csv('/work/longyh/BY/raw/SraRunTableWXS.csv')
patient_list=pd.read_csv('/work/longyh/BY/processed/s2_patient.txt',header=None,names=['Patient'])
```

```python
WES_meta.head()
```

```python
patient_list.head()
```

```python
WES_meta['Patient']=WES_meta['Sample Name'].str.split('_').str[0]
```

```python
WES_meta_filtered=WES_meta[WES_meta['Patient'].isin(patient_list['Patient'])]
```

```python
print(WES_meta.shape)
print(WES_meta_filtered.shape)
print(patient_list.shape)
```

```python
#过滤样本名带有_on后缀的
WES_meta_filtered=WES_meta_filtered[~WES_meta_filtered['Sample Name'].str.contains('_on')]
```

```python
WES_meta_filtered.to_csv('/work/longyh/BY/processed/WES_meta_filtered.csv',index=False)
```
